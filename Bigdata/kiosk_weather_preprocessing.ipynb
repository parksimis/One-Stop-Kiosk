```python
import pandas as pd
import numpy as np
```

### 파일 불러오기

```python
metadata = pd.read_csv("data/META_관측지점정보.csv", engine='python')
metadata
```

```python
data = pd.read_csv("data/OBS_ASOS_TIM_20210525100314.csv",encoding="UTF-8",parse_dates=["일시"])
data
```

### metadata 결측치 제거

```python
metadata.info()
```

```python
metadata = metadata[metadata["종료일"].isnull()]
```

### metadata 지역 컬럼 생성

```python
metadata["지역"] = metadata.loc[:, '지점주소'].str[:2]
```

```python
metadata
```

```python
pd.date_range('2021-05-16', '2021-05-17', freq='h')[:-1].shape[0] * data['지점명'].unique().shape[0]
```

###  metadata 와 날씨 데이터 merge

```python
weather_data = data.merge(metadata[["지점","지역"]], how="inner", left_on="지점", right_on="지점")
```

```python
weather_data
```

### 날씨 데이터의 지역 컬럼을 설문 데이터의 지역 컬럼 동일화

```python
weather_data["지역"] = np.where(weather_data["지역"]=="경상", '경상도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="부산", '경상도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="울산", '경상도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="대구", '경상도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="강원", '강원도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="전라", '전라도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="광주", '전라도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="인천", '수도권 (경기, 인천)',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="경기", '수도권 (경기, 인천)',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="세종", '충청도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="충청", '충청도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="대전", '충청도',weather_data["지역"])
weather_data["지역"] = np.where(weather_data["지역"]=="제주", '제주도',weather_data["지역"])
```

```python
weather_data
```

### 날씨 데이터를 지역, 일시를 기준으로 weather_mean생성 

```python
weather_mean = weather_data.groupby(["지역","일시"]).mean().reset_index()
weather_mean
```

### weather data 결측치 제거

```python
# 결측치 확인
weather_mean.isnull().sum()
```

```python
# 결측값 0으로 변경
weather_mean = weather_mean.fillna(0)
```

```python
# 전운량을 int타입으로 변경
weather_mean['전운량(10분위)'] = weather_mean['전운량(10분위)'].astype('int')
```

### 시계열데이터  년-월-일-시로 변경

```python
# 소수점 첫째자리까지 반올림
weather_mean.iloc[:, 3:-1] = round(weather_mean.iloc[:, 3:-1], 1)
weather_mean.iloc[:, -1:] = round(weather_mean.iloc[:, -1:])
```

```python
weather_mean.head(3)
```

```python
weather_mean.info()
```

```python
# 지점 삭제
weather_mean = weather_mean.drop(['지점'], axis=1)
```

```python
weather_mean.head(3)
```

```python
# 날짜
weather_mean['일시'] = pd.to_datetime(weather_mean['일시'])
```

```python
weather_mean.info()
```

```python
weather_mean["일시"] = weather_mean["일시"].dt.strftime('%Y-%m-%d-%H')
```
