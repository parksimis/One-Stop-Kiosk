```python
final_data = data.drop(['수집자 이름'], axis=1)
```

```python
tt = pd.get_dummies(final_data.iloc[:, 1:])
tt
```

```python
from sklearn.cluster import KMeans
sse = []

for i in range(1, 51):
    
    kmeans = KMeans(n_clusters=i) #세 개의 군집으로 분류
    kmeans.fit(tt)
    sse.append(kmeans.inertia_)
```

```python
x = [i for i in range(1, 51)]
plt.figure(figsize=(13, 13))
plt.plot(x, sse)
plt.scatter(x, sse)
plt.grid()
```

```python
print("Cluster memberships:\n{}".format(kmeans.labels_))
```

```python
final_data.columns
```

```python
from scipy.spatial.distance import pdist, squareform

row_dist = pd.DataFrame(squareform(pdist(kmeans.cluster_centers_, metric='euclidean')), columns=set(kmeans.labels_))
row_dist
```

```python
from scipy.cluster.hierarchy import dendrogram, ward

# Apply the ward clustering to the data array X
# The SciPy ward function returns an array that specifies the distances
# bridged when performing agglomerative clustering
linkage_array = ward(tt)
# Now we plot the dendrogram for the linkage_array containing the distances
# between clusters
plt.figure(figsize=(15, 10))
dn = dendrogram(linkage_array)
plt.xticks(rotation=90)
plt.show()
```
